```python
# %matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader.data as web
import pandas as pd
import datetime


start = datetime.datetime(1949, 1, 1)
end = datetime.datetime(2018, 5, 16)
tsd = web.DataReader('SP500','fred',start,end)
print(tsd.head(1))
print(tsd.tail(1))
sp500 = tsd.copy(1)

sp500.plot()
```

```python
vol = (np.log(sp500).diff().dropna().rolling(20).std()*np.sqrt(365)*100)
vol.plot()
```

```python
ts = pd.concat([vix,vol],axis=1)
ts.columns = ['vix','vol']
#
ts = ts.dropna(how='any')
ts.vix.plot(label='vid')
ts.vol.plot(label='vol')
plt.legend()
print(ts.vix.mean(),ts.vol.mean())
```
